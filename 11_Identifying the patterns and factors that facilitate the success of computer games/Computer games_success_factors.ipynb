# Project description

We work for Streamchik, an online shop which sells computer games worldwide. Historical data on game sales, user and expert ratings, genres and platforms (e.g. Xbox or PlayStation) are available from open sources. We need to identify the patterns that determine a game's success. This will allow us to bet on a potentially popular product and plan advertising campaigns.

We have the data in front of us up to 2016. Let's imagine it's December 2016 and we're planning a campaign for 2017. We need to work out how to work with the data. It doesn't matter if we are forecasting sales for 2017 based on 2016 data or 2027 based on 2026 data.

In the data set comes the abbreviation ESRB (Entertainment Software Rating Board), which is an association that determines the age rating of computer games. The ESRB evaluates game content and assigns it an appropriate age category, such as "For Adults", "For Young Children" or "For Teens".

* Rating - is a rating from the ESRB (Entertainment Software Rating Board). This association determines the rating of computer games and assigns them an appropriate age category.

## Data preprocessing

### Changing the case of the column names

In addition, right away the data in the columns with the text data type will be lowercased.

```python
# lowercasing the columns names

df.columns = df.columns.str.lower()

# lowercasing the values in respective columns

for column in ['name', 'platform', 'genre', 'rating']:
    df[column] = df[column].str.lower()
```

### Converting data to correct data types

Converting the data in the `year_of_release` and `user_score` columns to the correct data types. 

**- `year_of_release` column**

Before converting the data type in the `year_of_release` column, it is necessary to handle the missing values, as the transformation is not possible if they are present. There are 269 missing values in this column (1.6%). The number is large enough to fill in manually, but insignificant when it is removed from the dataset. Removing these missing data will not distort our dataset too much. Thus, we delete these missing data and then convert the type of data in the column. 

Despite the fact that the column contains only year (without month and day), we will not convert it to date data type, as it is not necessary. Instead we convert it to an integer type.

```python
df = df.dropna(subset=['year_of_release']).reset_index(drop=True)
df.year_of_release = df.year_of_release.astype(int)
```

**- `user_score' column**

Before converting the data type in the user ratings column, it is necessary to handle the most frequent text value in this column, "tbd". because with its presence the transformation is not possible. 

There are 2424 "tbd" values in this column. In a Google search for "tbd in user score", the first line of the output reads:

>">What does "tbd" mean? To Be Determined. To help ensure that METASCORES accurately reflect the reviews given by critics for any particular movie, game, television show or album, we do not display a METASCORE for those items that do not have at least four (4) reviews in our database.*"

That is, the value "tbd" means that the user rating is not calculated because there are at less than four ratings for a given game. Consequently, this value is not an error in the data and should be saved.

In order to convert to the `float` data type, this value should be replaced by a numeric placeholder that does not lie between 0 and 10 (true user scores). For example, replace it with a value of "-1". 

```python
df.user_score.replace('tbd', -1, inplace=True)
df.user_score = df.user_score.astype(float)
```

---
**- columns `critic_score` & `user_score`**

There are a very large number of missing values in the critics' and users' assessment data columns. It is necessary to understand the nature of these omissions in order to determine the approach to filling them in or not.

Let's briefly check the distribution of scores by year of game release. We will assess both columns together, as the nature of the data and the missing values in these columns are likely to be similar.

```python
df.groupby('year_of_release').agg({'critic_score':'count', 'user_score':'count'}).plot(grid=True)
```

It can be seen that before 1995 - before the Internet era - there are no assessments of both critics and users. Most likely the missing values are due to the fact that before that period no assessments were collected at all. Let us check this assumption.

```python
print('ÐŸÑ€Ð¾Ð¿ÑƒÑÐºÐ¸ Ð¿Ð¾ÑÐ»Ðµ 1995 Ð³Ð¾Ð´Ð° Ñƒ ÐºÑ€Ð¸Ñ‚Ð¸ÐºÐ¾Ð²:', len(df[(df['critic_score'].isna()==True)&(df['year_of_release']>1995)]))
print('ÐŸÑ€Ð¾Ð¿ÑƒÑÐºÐ¸ Ð¿Ð¾ÑÐ»Ðµ 1995 Ð³Ð¾Ð´Ð° Ñƒ Ð¿Ð¾Ð»ÑŒÐ·Ð¾Ð²Ð°Ñ‚ÐµÐ»ÐµÐ¹:', len(df[(df['user_score'].isna()==True)&(df['year_of_release']>1995)]))
```

Our assumption turned out to be wrong and the missing values in these columns are not concentrated only in the period before 1995 - before the era of the development of the Internet. More precisely, in this period there are also missing data, but there are too few records for this period (only 486 observations), so there are few missing data. But in any case, the presence of omissions in this time period is quite logical, as it shows that in that period it was not physically possible to give marks by critics (there were no specialized Internet sites) and by users.

As for data after - 1996, the newly appearing publications that publish game reviews and leave reviews and ratings, do not tend to review everything in a row: it makes no sense, and besides there are too many games coming out. The high lack of reviews for most games is therefore not surprising. There are quantitatively more users than specialized magazines - so there are fewer omissions and more scores, which is clearly visible in the graph. 

The lack of data in these columns is, in our view, quite natural and filling them in does not seem logical or necessary. Moreover, for our study of game sales, filling in the missing data in these columns is insignificant.  

The only thing is that since the gaps indicate the real absence of score for each particular game, we can replace them with a numerical placeholder outside the interval from 0 to 100 (the true scores of users and critics). For example, a value of "-1", with which we have also replaced "tbd".

```python
df.critic_score.fillna(-1, inplace=True)
df.user_score.fillna(-1, inplace=True)
```

## Hypothesis testing

### Average user ratings for Xbox One and PC platforms are the same

Let us test the Hypotheses of Equality between the two General Sets. To do this, we first select two samples for the actual period with a compared indicator of user rating, in the context of the corresponding game platforms. 

Let us form two samples for which we will conduct statistical tests. We will also calculate the variance for them in order to apply the statistical test correctly later on.

```python
xone = df.query('2011<year_of_release and platform=="xone" and user_score!=-1').user_score
pc = df.query('2011<year_of_release and platform=="pc" and user_score!=-1').user_score

print("Ð”Ð¸ÑÐ¿ÐµÑ€ÑÐ¸Ñ XONE:", xone.var())
print("Ð”Ð¸ÑÐ¿ÐµÑ€ÑÐ¸Ñ PC:", pc.var())
```

We need to test the hypothesis that the average user scores of the XONE and PC platforms are the same. When testing this hypothesis, it should be taken into account that our samples are independent of each other. Therefore we will use `ttest_ind()` to test the hypotheses.

In addition, the variance values calculated above in terms of platforms suggest that, when applying the test, the variance difference will need to be specified in the `equal_var = False` parameter of the `ttest_ind()` method.

Let us formulate the null hypothesis and the alternative hypothesis, given that the null hypothesis always implies equality of the compared indicators, and the alternative hypothesis, respectively, complete or one-sided inequality:

- H0: "The average user scores of the XONE and PC platforms are the same".
- H1: "The average user ratings of XONE and PC platforms are different".

Assume a critical level of statistical significance `alpha = 0.05`. If `p-value` is less than it, we reject our null hypothesis.

```python
alpha = 0.05

results = st.ttest_ind(xone, pc, equal_var = False)

print('p-value =', results.pvalue)

if results.pvalue < alpha:
    print("ÐžÑ‚Ð²ÐµÑ€Ð³Ð°ÐµÐ¼ Ð½ÑƒÐ»ÐµÐ²ÑƒÑŽ Ð³Ð¸Ð¿Ð¾Ñ‚ÐµÐ·Ñƒ")
    
else:
    print("ÐÐµ Ð¿Ð¾Ð»ÑƒÑ‡Ð¸Ð»Ð¾ÑÑŒ Ð¾Ñ‚Ð²ÐµÑ€Ð³Ð½ÑƒÑ‚ÑŒ Ð½ÑƒÐ»ÐµÐ²ÑƒÑŽ Ð³Ð¸Ð¿Ð¾Ñ‚ÐµÐ·Ñƒ")
```

The result of the hypothesis test shows that we were unable to reject our null hypothesis. 

**That is, the average user scores of the XONE and PC platforms are not statistically significantly different.** That is, there is no statistically significant difference between the samples.

### Average user ratings for the Action and Sports genres are different

Let us test the Hypotheses of Equality between the two General Sets. To do this, we first select two samples for the actual period with a compared indicator of user rating, in the context of the corresponding game platforms.
